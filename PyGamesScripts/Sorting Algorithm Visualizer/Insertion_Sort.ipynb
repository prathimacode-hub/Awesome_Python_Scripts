```python
# Importing modules

import pygame
import random

pygame.font.init()

screen = pygame.display.set_mode((700, 700))

pygame.display.set_caption("Insertion Sort")

# bool variable to run the sorting visualzation, as long as this is true
run = True

width = 700
length = 600
array =[0]*50
arr_color =[(0, 204, 102)]*50

color =[(0, 255, 0), (255, 0, 0), (0, 0, 255), (255, 102, 0)]

fnt = pygame.font.SysFont("helvetica", 30)

# generating array
def generate_arr():
    for i in range(1, 50):
        arr_color[i]= color[0]
        array[i]= random.randrange(1, 100)

generate_arr()

# we have to refill the screen to keep the sorting continued
def refill():
    screen.fill((0, 0, 0))
    draw()
    pygame.display.update()
    pygame.time.delay(10)

# sorting algorithm
def insertionSort(array):

    for i in range(1, len(array)):
        pygame.event.pump()
        refill()
        key = array[i]
        arr_color[i]= color[2]
        j = i-1
        while j >= 0 and key < array[j]:
            arr_color[j]= color[2]
            array[j + 1]= array[j]
            refill()
            arr_color[j]= color[3]
            j = j-1
        array[j + 1]= key
        refill()
        arr_color[i]= color[0]

# to draw the array elements as bars
def draw():
    txt = fnt.render("PRESS ENTER", 1, (0, 0, 0))
    # positioning of the above text
    screen.blit(txt, (230, 20))
    element_width =(width-50)//50
    boundryArr = 700 / 50
    boundryGrp = 550 / 100
    pygame.draw.line(screen, (0, 0, 0), (0, 95), (700, 95), 6)
    
    for i in range(1, 50):
        pygame.draw.line(screen, arr_color[i], (boundryArr * i-3, 100), (boundryArr * i-3, \
                                                                            array[i]*boundryGrp + 100), element_width)

while run:
    screen.fill((100, 100, 100))
    # event handler for pygame
    for event in pygame.event.get():

        if event.type == pygame.QUIT:
            run = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_RETURN:
                insertionSort(array)    
    draw()
    pygame.display.update()
    
pygame.quit()
```
